```python
#Import pandas to read csv files and create dataframe from the csv file
#Import matplotlib to plot the graph on the final data 
%matplotlib inline
import pandas as pd
#import seaborn,matplotlib and numpy libray

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#filename is variable to store the fully qualified path of titanic-data.csv
#Read the titanic-data.csv file using pandas and get the result in dataframe titanic_data
filename = 'C:/Users/KHAN/Anaconda2/DataFiles/titanic-data.csv'
titanic_data = pd.read_csv(filename)
#Print top 5 rows of the data frame to check the data
titanic_data[:5]
```

#Add a new column in the data frame for Survival Status. Values 'Survived' means the passenger survived, 
#'Deceased' means passenger didn't survive.

```python
titanic_data.loc[titanic_data['Survived'] == 1, 'Survival_Status'] = 'Survived';
titanic_data.loc[titanic_data['Survived'] == 0, 'Survival_Status'] = 'Deceased';
#print top 5 rows of the updated data set
titanic_data[:5]
```

#Add a new column in the data frame for Class Type. Values '1st', '2nd' & '3rd' means the passenger was travelling 
#in first, second or third class and 'Nan' if the value is missing in csv file 

```python
titanic_data.loc[titanic_data['Pclass'] == 1, 'Class_Type'] = '1st';
titanic_data.loc[titanic_data['Pclass'] == 2, 'Class_Type'] = '2nd';
titanic_data.loc[titanic_data['Pclass'] == 3, 'Class_Type'] = '3rd';
#print top 5 rows of the updated data set
titanic_data[:5]
```

#Add a new column in the data frame for AgeGroup. Values in this column will be Child(age >0 and age <=14), 
#Adult(age>14 and age<=60), Elder(age>60) and _NaN(means value is missing in csv file)
#Age values is 177 passengers, this makes it difficult to calculate how many out of 177 were child, adult or elder
#But, this analyis will help in caldulating how many of 177 were survived

```python
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=[0,14,60,90], labels=['Child','Adult','Senior'])
#print top 5 rows of the updated data set
titanic_data[:5] 
```

#Add 2 new columns Survival_Yes and Survival_No in the data frame for Survival_Status. 
#Values in this column will be 0 or 1 based on if Survival_Status of the passenger 
#Added these 2 columns mainly because I want 2 separate columns having sum/count of passengers survived or not.
#Final graph will be plotted against these 2 columns.

```python
titanic_data.loc[titanic_data['Survival_Status'] == 'Survived', 'Survival_Yes'] = 1;
titanic_data.loc[titanic_data['Survival_Status'] == 'Survived', 'Survival_No'] = 0;
titanic_data.loc[titanic_data['Survival_Status'] == 'Deceased', 'Survival_Yes'] = 0;
titanic_data.loc[titanic_data['Survival_Status'] == 'Deceased', 'Survival_No'] = 1;
#print top 5 rows of the updated data set
titanic_data[:5] 
```

Add a function DrawGraph to plot the similar graph instead of having the same code at multiple places for similar graphs
This function is used to plot the bar graphs. 
Below are the parameters with description:
df - dataset on which graphs should be plotted
graph_type - which graph type to be plotted e.g. bar
title - title of the graph
xlabel - label on x axis of the graph
ylabel - lebel on y axis of the graph
graph_size - size of the graph in term of width, height

```python
def DrawGraph(df,graph_type,title,xlabel, ylabel, graph_size):
    if(df.empty):
        print("Dataframe is empty")
    else:
        ax = df.plot(kind=graph_type,title=title,figsize=(graph_size))
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
```

#Plot the graph for the number of passengers survived/didn't survive against Groupby Sex, AgeGroup, Class_Type 
#Below graph shows the maled vs female survival rate, other parameters taken into account were male/female, age group and 
#the class of passengers. Survival count of passengers are plotted against three variables sex, agegroup & class of travel 
#for quick and easy comparison for the number of passengers survived or not survived in each group.

```python
df = titanic_data.groupby(['Sex','AgeGroup','Class_Type'])['Survival_Yes','Survival_No'].sum()
DrawGraph(df,'bar','Passengers Survival/deceased count vs Sex/AgeGroup/Class','Sex/AgeGroup/Class','Count',(16, 4))
```

#Plot the graph for the number of passengers survived/didn't survive against Groupby Sex, AgeGroup
#Below graph shows the maled vs female survival rate, another parameter taken into account is in which age group they belong to. 
#The below graphs shows the survival rate of male/female based on their age group e.g. Child, Adult, Elder.
#Values of age were missing for many passengers, for those, a new age group '_NaN' was created.
#and displayed the surivial rate of those male/female passengers accordingly. 
#The below graph quickly compares the survival rate of male vs female passengers based on their age group.

```python
df = titanic_data.groupby(['Sex','AgeGroup'])['Survival_Yes','Survival_No'].sum()
DrawGraph(df,'bar','Passengers Survival/deceased count vs Sex/AgeGroup','Sex/AgeGroup','Count',(16, 4))
```

#Plot the graph for the number of passengers survived/didn't survive against Groupby Sex, Class_Type 
#Below graph shows the male vs female survival rate, another parameter taken into account is in which classes different 
#different male & female passengers were travelling and their survival rate from each class category. 

```python
df = titanic_data.groupby(['Sex','Class_Type'])['Survival_Yes','Survival_No'].sum()
DrawGraph(df,'bar','Passengers Survival/deceased count vs Sex/Class','Sex/Class','Count',(16, 4))
```

```python
#Plot the factorplot with columns of Class_Type, age on y axis, male/female survived/not-survived on x-axis 
#Below graph shows more female survived in 1st, 2nd & 3rd class

sns.factorplot(data=titanic_data, col='Class_Type', x='Sex', y='Age', hue='Survival_Status', 
               kind='swarm', size=10, aspect=0.5, s=10, legend_out=True)
plt.yticks(np.arange(0,81,10))
plt.subplots_adjust(top=0.9)
plt.suptitle('Survived/Deceased male and female passengers travelling in 1st, 2nd & 3rd class vs Age')
plt.show()
```

```python
#Plot the barplot, the below graphs shows the survival ratio of male & female travelling in different classes 
#i.e. 1st, 2nd & 3rd 
plt.subplots(figsize=(16,4))
sns.barplot(data=titanic_data, x="Sex", hue='Class_Type', y='Survived', estimator=np.mean);
plt.ylabel('Percentage of Survival');
plt.title('Relative Survival Rate of passengers in 1st, 2nd & 3rd class vs Male/Female')
plt.show()
```
